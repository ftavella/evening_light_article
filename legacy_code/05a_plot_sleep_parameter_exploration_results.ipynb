```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
selected_parameters_early = pd.read_csv('data/sleep_parameter_exploration/early_light_selected_sleep_parameters.csv')
selected_parameters_late = pd.read_csv('data/sleep_parameter_exploration/late_light_selected_sleep_parameters.csv')
```

```python
print(f'Length of early light selected parameters: {len(selected_parameters_early)}')
print(f'Length of late light selected parameters: {len(selected_parameters_late)}')
```

```python
# Histogram of sleep times and bedtime
fig, ax = plt.subplots(1, 3, figsize=(11, 3))

ax[0].hist(selected_parameters_early['sleep_time'], bins=20, color='tab:blue', 
              alpha=0.7, rwidth=0.8, label=f'6 AM - 10 PM Light (N={len(selected_parameters_early)})',
              range=(6.5, 8.0))
ax[0].hist(selected_parameters_late['sleep_time'], bins=20, color='tab:red', 
              alpha=0.7, rwidth=0.8, label=f'8 AM - 12 AM Light (N={len(selected_parameters_late)})',
              range=(6.5, 8.0))
ax[0].set_xlabel('Sleep duration (hours)')
ax[0].set_ylabel('Counts')
ax[0].legend(fontsize='small', ncols=2, 
             bbox_to_anchor=(1.75, 1.1), loc='center')

ax[1].hist(selected_parameters_early['bedtime'], bins=20, color='tab:blue', 
           alpha=0.7, rwidth=0.8, range=(21.5, 26))
ax[1].hist(selected_parameters_late['bedtime'], bins=20, color='tab:red', 
           alpha=0.7, rwidth=0.8, range=(21.5, 26))
ax[1].set_xlabel('Bedtime')
ax[1].set_ylabel('Counts')
ax[1].set_xticks([22, 23, 24, 25])
ax[1].set_xticklabels(['10 PM', '11 PM', '12 AM', '1 AM'])

ax[2].hist(selected_parameters_early['dlmo'], bins=20, color='tab:blue',
              alpha=0.7, rwidth=0.8, range=(20, 23))
ax[2].hist(selected_parameters_late['dlmo'], bins=20, color='tab:red',
                alpha=0.7, rwidth=0.8, range=(20, 23))
ax[2].set_xlabel('DLMO')
ax[2].set_ylabel('Counts')
ax[2].set_xticks([20, 21, 22, 23])
ax[2].set_xticklabels(['8 PM', '9 PM', '10 PM', '11 PM'])

fig.subplots_adjust(wspace=0.25)

plt.savefig('figures/sleep_parameter_exploration_sleep_time_bedtime.pdf', bbox_inches='tight')

plt.show()

```
