```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_path = 'data/light_schedules/'
```

```python
time = np.load(f'{data_path}/time.npy')
regular_early_light = np.load(f'{data_path}/regular_early_light.npy')
regular_late_light = np.load(f'{data_path}/regular_late_light.npy')
chang14_ebook_first_light = np.load(f'{data_path}/chang14_ebook_first_light.npy')
chang14_ebook_second_light = np.load(f'{data_path}/chang14_ebook_second_light.npy')
chang14_high_ebook_first_light = np.load(f'{data_path}/chang14_high_ebook_first_light.npy')
chang14_high_ebook_second_light = np.load(f'{data_path}/chang14_high_ebook_second_light.npy')
realistic_ebook_first_light = np.load(f'{data_path}/realistic_ebook_first_light.npy')
realistic_ebook_second_light = np.load(f'{data_path}/realistic_ebook_second_light.npy')
```

## Visualize all schedules

```python
time_mask = (time >= 48) & (time <= 72)
relative_time = time[time_mask] - 48

y_options = [
    regular_early_light[time_mask],
    regular_late_light[time_mask],
    chang14_ebook_first_light[time_mask],
    chang14_ebook_second_light[time_mask],
    chang14_high_ebook_first_light[time_mask],
    chang14_high_ebook_second_light[time_mask],
    realistic_ebook_first_light[time_mask],
    realistic_ebook_second_light[time_mask]
]

y_labels = [
    'Regular (early)',
    'Regular (late)',
    'Chang14 ebook first',
    'Chang14 ebook second',
    'Chang14 high first',
    'Chang14 high second',
    'Realistic ebook first',
    'Realistic ebook second'
]

for i in range(7):
    fig, ax = plt.subplots()

    print(f'Light values: {np.unique(y_options[i])}')
    ax.plot(relative_time, y_options[i], label=y_labels[i])

    ax.set_xlabel('Time (h)')
    ax.set_ylabel('Light (lux)')
    ax.set_yscale('symlog')
    ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.show()

```

## Article viz

```python
# Chang 14
lw = 2
time_mask = (time >= 48) & (time <= 72)
relative_time = time[time_mask] - 48
book_light = chang14_ebook_second_light[time_mask]
ebook_light = chang14_ebook_first_light[time_mask]

fig, ax = plt.subplots(figsize=(3.5,2))
ax.plot(relative_time, book_light, color='tab:blue', 
        zorder=1, linewidth=lw, label='Book')
ax.plot(relative_time, ebook_light, color='tab:red', 
        zorder=0, linewidth=lw, label='eBook')
ax.axvspan(18, 22, facecolor='gray', alpha=0.3,
           edgecolor='none', zorder=-1,
           label='Reading')
        
ax.set_xlabel('Time (h)')
ax.set_ylabel('Light (lux)')
ax.set_ylim(-2, 100)
ax.legend(bbox_to_anchor=(0.01, 1.2), loc='upper left', 
          borderaxespad=0., ncol=3, fontsize='small')
plt.savefig('figures/chang14_ebook_light_schedule.pdf',
            bbox_inches='tight')

plt.show()
```
